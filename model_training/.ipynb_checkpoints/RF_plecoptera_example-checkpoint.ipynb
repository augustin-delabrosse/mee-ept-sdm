```python
import pandas as pd
import numpy as np
import geopandas as gpd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectFromModel
from sklearn import metrics
import matplotlib.pyplot as plt
import shap

from config import * # Custom configuration file
```

# Load data

```python
test_cisse = gpd.read_file(split_result_stickies + f"cisse_stat_test/" + f"sticky_cisse_test.shp")
test_louroux = gpd.read_file(split_result_stickies + f"louroux_stat_test/" + f"sticky_louroux_test.shp")
test_fao = gpd.read_file(split_result_stickies + f"fao_stat_test/" + f"sticky_fao_test.shp")
test_roudoudour = gpd.read_file(split_result_stickies + f"roudoudour_stat_test/" + f"sticky_roudoudour_test.shp")
test_timbertiere = gpd.read_file(split_result_stickies + f"timbertiere_stat_test/" + f"sticky_timbertiere_test.shp")
```

```python
train_cisse = gpd.read_file(split_result_stickies + f"cisse_stat_train/" + f"sticky_cisse_train.shp")
train_louroux = gpd.read_file(split_result_stickies + f"louroux_stat_train/" + f"sticky_louroux_train.shp")
train_fao = gpd.read_file(split_result_stickies + f"fao_stat_train/" + f"sticky_fao_train.shp")
train_roudoudour = gpd.read_file(split_result_stickies + f"roudoudour_stat_train/" + f"sticky_roudoudour_train.shp")
train_timbertiere = gpd.read_file(split_result_stickies + f"timbertiere_stat_train/" + f"sticky_timbertiere_train.shp")
```

```python
concat_train = pd.concat([train_cisse, train_louroux, train_fao, train_roudoudour, train_timbertiere])
concat_test = pd.concat([test_cisse, test_louroux, test_fao, test_roudoudour, test_timbertiere])
```

```python
# === Select relevant features ===
# Features include environmental variables and insect counts
cols = ['Nb_Ephemer', 'Nb_Plecopt', 'Nb_Trichop',  # Insect counts
       'd2water', 'd2woods',  # Distance to water and forest
       'wd_5m', 'wd_25m', 'wd_50m',  # Vegetation density at different distances
       'imean_5m', 'ivar_5m', 'imean_25m', 'ivar_25m', 'imean_50m', 'ivar_50m',  # Intensity mean & variance
       'emean_5m', 'evar_5m', 'emean_25m', 'evar_25m', 'emean_50m', 'evar_50m',  # Elevation mean & variance
       'elev', 'geometry']

feature_cols = cols[3:-1]  # Selecting only numerical features
```

```python
# Remove rows with missing values
train_points = concat_train[cols]
train_points = train_points.dropna()

test_points = concat_test[cols]
test_points = test_points.dropna()
```

```python
train_points.describe()
```

```python
test_points.describe()
```

# Preprocess data

```python
# === Define the target variable ===
y_col = "Nb_Plecopt"  # Predicting the presence of Trichoptera (caddisflies)

# Split the dataset randomly into training (80%) and testing (20%)
train, test = train_points.sample(frac=1).reset_index(), test_points

# Extract features (X) and target variable (y)
X_train = train[feature_cols]
y_train = 1*(train[y_col] > 0) # Convert to binary classification
X_test = test[feature_cols]
y_test = 1*(test[y_col] > 0)
```

```python
print(f"""dataset d'entraînement : 
{y_train.value_counts()/len(y_train)}

dataset de test : 
{y_test.value_counts()/len(y_test)}""")
```

```python
# === Normalize the features ===
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# === Feature Selection using Random Forest  ===
selector = SelectFromModel(RandomForestClassifier(n_estimators=10000), threshold="median")
selector.fit(X_train_scaled, y_train)

# Transform the dataset using selected features
X_train_selected = selector.transform(X_train_scaled)
X_test_selected = selector.transform(X_test_scaled)
```

# Define and train Random Forest

```python
# Retrain Random Forest with selected features
rf = RandomForestClassifier(n_estimators=10000)
rf.fit(X_train_selected, y_train)
```

# Evalute model

```python
# Predictions and evaluation
y_pred = rf.predict(X_test_selected)
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(metrics.classification_report(y_test, y_pred))#, target_names=label_encoder.classes_))
```

```python
# === Compute and plot ROC curve ===
y_proba = rf.predict_proba(X_test_selected)
fpr, tpr, thresholds = metrics.roc_curve(y_test, [i[1] for i in y_proba])
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                  estimator_name='example estimator')
display.plot()
plt.show()
```

# Get most important features with SHAP

```python
# === Explain model predictions using SHAP ===
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_train_scaled)
```
